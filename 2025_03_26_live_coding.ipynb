###Visualize Sales Trends
Create time series plots to visualize sales trends over time for each region. Highlight any noticeable patterns or seasonal variations in the sales data.

Time series plots are created to visualize and analyze sales trends over time for each region.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df['Date'], df['Revenue'], marker="x") # line plot
ax.set_xlabel('Date')
ax.set_ylabel('Revenue')
ax.set_title('Sales Trends')
plt.show()
```

### 3) Compare Product Performance
Use Seaborn to create comparative visualizations that analyze the performance of different product categories across regions. Consider visualizations like bar plots, box plots, or violin plots for this analysis.

Comparative visualizations are created to analyze product performance across regions.

```python
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Region', y='Revenue', hue='Product')
plt.title('Product Performance by Region')
plt.xlabel('Region')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.violinplot(data=df, x='Region', y='Revenue', hue='Product')
plt.title('Product Performance by Region')
plt.xlabel('Region')
plt.show()
```

### Perform Correlation Analysis
Investigate the relationship between sales figures and other numerical variables in the dataset. Use Seaborn to create heatmaps showing the correlation matrix of the dataset.

Correlation analysis is performed, and results are visualized using heatmaps.

```python
df_numerical=df.select_dtypes(include=['number'])
df_numerical
```

```python
corr = df_numerical.corr()
corr
```

```python
plt.figure(figsize=(20, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
