Create a function named get_longest_movie() that determines the movie with the longest runtime from a list of movies provided as an argument.
A function called get_longest_movie() is defined, returning the movie with the longest runtime from a provided list of movies.

```python
def get_longest_movie(filmy):
    run_time=0
    title=""
    for film in filmy:
        if film["runtime"]>run_time:
            run_time=film["runtime"]
            title=film["title"]
    return title, run_time
print(get_longest_movie(movies))
```

```python
def pocet_nul(a):  # definování funkce pocet nul v listu
    pocet = 0
    for i in a:
        if i == 0:
            pocet += 1
    return pocet
pocet_nul([0,0,1,2])
```

```python
def pocet_nul2(a):  # definování funkce pocet nul v listu
    pocet2=[1 for i in a if i == 0]
    print(sum(pocet2))

pocet_nul2([0,0,1,2])
```

```python
%%timeit
def delitelne_tremi(a):  # definování funkce pocet čísel dělitelných třemi
    pocet =  []
    for i in a:
        if i % 3 == 0:
            pocet.append(i)
    return pocet

print(delitelne_tremi([1,3,5,6,9]))
```

```python
%%timeit
def delitelne_tremi2(a):
    pocet2=[i for i in a if i % 3 == 0]
    return pocet2

print(delitelne_tremi2([1,3,5,6,9]))
```
